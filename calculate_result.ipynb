```python

#for K in [20,25]:


#print(thing)
#print(len(thing))
        
for name in EXPERIMENT_DIRS:

    model_line = []
    forecast_line = []
    airnow_line = []

    model_cleaned = []
    forecast_cleaned = []
    airnow_cleaned = []

    #print("Clusters for K="+str(K)+":")
    print(name+":")
    
    for i in range(K):
        #'plots/K'+str(K)+'/K'+str(K)+'_'+str(i)+'/data.p'
        #(TEST_LOCATIONS, TEST_DAYS, PHYSICAL_VARIABLES, out, rmse, rmse1, pearson, max1, max2, model_output, weights) = pickle.load(open('plots/K'+str(K)+'/'+str(i)+'/data.p', 'rb'))
        (TEST_LOCATIONS, TEST_DAYS, PHYSICAL_VARIABLES, out, rmse, rmse1, pearson, max1, max2, model_output, weights) = pickle.load(open('plots/'+name+'/'+str(i)+'/data.p', 'rb'))

        model_line.append(np.concatenate(out['model']))
        forecast_line.append(np.concatenate(out['forecast']))
        airnow_line.append(np.concatenate(out['airnow']))
        
        for j,x in enumerate(airnow_line[i]):
            if airnow_line[i][j]<-100 or model_line[i][j]<-100:
                airnow_line[i][j] = np.NaN

        model_cleaned.append(np.delete(model_line[i],np.argwhere(np.isnan(airnow_line[i]))))
        forecast_cleaned.append(np.delete(forecast_line[i],np.argwhere(np.isnan(airnow_line[i]))))
        airnow_cleaned.append(np.delete(airnow_line[i],np.argwhere(np.isnan(airnow_line[i]))))

        print("Cluster "+str(i)+":")
        
        diff = list(model_cleaned[i] - airnow_cleaned[i])
        index = min(range(len(diff)), key=diff.__getitem__)
        print(model_cleaned[i][index])
        #print(min(list(model_cleaned[i] - airnow_cleaned[i])))
        
        print("Forecast RMSE:",mean_squared_error(forecast_cleaned[i], airnow_cleaned[i], squared=False))
        print("Model RMSE:",mean_squared_error(model_cleaned[i], airnow_cleaned[i], squared=False))
        print("Reduction:",mean_squared_error(forecast_cleaned[i], airnow_cleaned[i], squared=False)-mean_squared_error(model_cleaned[i], airnow_cleaned[i], squared=False))
        print()

    #print("Overall for K="+str(K)+":")
    print("Overall for "+name+":")
    model_line_all = np.concatenate(model_line)
    forecast_line_all = np.concatenate(forecast_line)
    airnow_line_all = np.concatenate(airnow_line)

    model_cleaned_all = np.delete(model_line_all,np.argwhere(np.isnan(airnow_line_all)))
    forecast_cleaned_all = np.delete(forecast_line_all,np.argwhere(np.isnan(airnow_line_all)))
    airnow_cleaned_all = np.delete(airnow_line_all,np.argwhere(np.isnan(airnow_line_all)))

    print("Forecast RMSE:",mean_squared_error(forecast_cleaned_all, airnow_cleaned_all, squared=False))
    print("Model RMSE",mean_squared_error(model_cleaned_all, airnow_cleaned_all, squared=False))
    print("Reduction:",mean_squared_error(forecast_cleaned_all, airnow_cleaned_all, squared=False)-mean_squared_error(model_cleaned_all, airnow_cleaned_all, squared=False))
    print()
```
